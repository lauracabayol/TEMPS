```python
weight_dict={(-99,0.99):0,
             (1,1.99):0.5,
             (2,2.99):0.75,
             (3,4):1,
             (9,9.99):0.25,
             (10,10.99):0,
             (11,11.99):0.5,
             (12,12.99):0.75,
             (13,14):1,
             (14.01,40):0
            }
```

```python
def map_weight(Qz):
    for key, value in weight_dict.items():
        if key[0] <= Qz <= key[1]:
            return value
    return None

# Apply the function to create the 'wQz' column
cat_calib['w_Q_f_S15'] = cat_calib['Q_f_S15'].apply(map_weight)
```
